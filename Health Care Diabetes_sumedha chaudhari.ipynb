# Healthcare
Problem Statement 

NIDDK (National Institute of Diabetes and Digestive and Kidney Diseases) research creates knowledge about and treatments for the most chronic, costly, and consequential diseases.

The dataset used in this project is originally from NIDDK.

The objective is to predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. 

Build a model to accurately predict whether the patients in the dataset have diabetes or not.



<b>Dataset Description</b>

The datasets consists of several medical predictor variables and one target variable (Outcome). Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and more.

- Pregnancies: Number of times pregnant

- Glucose: Plasma glucose concentration a 2 hours in an oral glucose tolerance test

- BloodPressure: Diastolic blood pressure (mm Hg)

- SkinThickness: Triceps skin fold thickness (mm)

- Insulin: 2-Hour serum insulin (mu U/ml)

- BMI: Body mass index (weight in kg/(height in m)^2)

- DiabetesPedigreeFunction: Diabetes pedigree function

- Age: Age (years)

- Outcome: Class variable (0 or 1)

# Correlation
Correlation explains how one or more variables are related to each other.If two variables are closely correlated, then we can predict one variable from the other. There are three types of correlation as follows:

Positive correlation -If the correlation value is> 0, there is a positive correlation. While the value of one variable increases, the value of the other variable also increases.

No Correlation- Correlation = 0 means no correlation.

Negative correlation- If the correlation is <0, there is a negative correlation. While one variable increases, the other variable decreases.

# Model Building 




```python
features=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
X=df[features]
y= df.Outcome
```

### Applying feature scaling

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(X)
sc_X=scaler.transform(X)
```

### Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(sc_X,y,test_size=0.2,random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Classification Algorithms

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
#Prediction
model_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of training data:",(model.score(X_train,y_train)*100))
print("Accuracy of test data:",(accuracy_score(y_test,model_predict)*100))
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix
cm = confusion_matrix (y_test, model_predict)
print(cm)
```

```python
plot_confusion_matrix(model,X_test,y_test,cmap="Blues",values_format=".3g")
```

```python
#Classification report
from sklearn import metrics
print("Classification report :\n")
print(classification_report(y_test,model_predict))
print("\nPrecision:",metrics.precision_score(y_test, model_predict))
print("\nRecall:",metrics.recall_score(y_test,model_predict))
```

```python
#AUC ROC curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
lr_prob=model.predict_proba(X_test)
lr_prob1=lr_prob[:,1]
fpr,tpr,thresh=metrics.roc_curve(y_test,lr_prob1)
roc_auc_lr=metrics.auc(fpr,tpr)
plt.figure(dpi=80)
plt.title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr,tpr,'y',label='AUC Score = %0.2f'%roc_auc_lr)
plt.plot(fpr,fpr,'r--',color='blue')
plt.legend()
plt.grid()
```
